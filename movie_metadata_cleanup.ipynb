```python
# Create a df with the most popular movies containing state names in the title for each state

# import the state names from the resources folder
states_db = pd.read_csv("resources/us_states.csv", quotechar='"', escapechar='\\')
states_db.head()
```

```python
# iterate through each state and find the movie title containing that state
# which is the most popular

states_by_titles_df = pd.DataFrame(columns=['state', 'movie_title'])
for i, row in states_db.iterrows():
    current_state = row['state']
    best_title = ''
    p = '\s' + current_state
    movie_titles = movie_summary_clean_df[movie_summary_clean_df['title'].str.contains(p, case=False)].sort_values(by='popularity', ascending=False)
    if movie_titles.empty:
        best_title = 'blank'
    else:
        best_title = movie_titles.iloc[0]['title']
    new_row = pd.DataFrame({'state': [current_state], 'movie_title': [best_title]})
    states_by_titles_df = pd.concat([states_by_titles_df, new_row], ignore_index=True)
print(states_by_titles_df)
```

```python
# check how many blank movie titles there are
states_by_titles_df['movie_title'].value_counts()
```

```python
# use the movie titles to get the rows from movie_summary_clean_df
filtered_states_by_titles_df = states_by_titles_df[states_by_titles_df['movie_title'] != 'blank']
state_title_movies_df = movie_summary_clean_df.loc[movie_summary_clean_df['title'].isin(filtered_states_by_titles_df['movie_title'])]
state_title_movies_df.head()
```

```python
movie_summary_overview_df = raw_db[['title', 'release_date', 'runtime', 'budget', 'revenue', 'popularity',
                                  'vote_average', 'vote_count', 'overview']]
movie_summary_overview_df.head(10)
```

```python
movie_summary_overview_df.dropna(subset=['popularity'], inplace=True)
movie_summary_overview_df.info()
```

```python
# clean the summary overview dataframe just enough to search through the overview column
# and sort by popularity

movie_summary_overview_df.dropna(subset=['title'], inplace=True)
movie_summary_overview_df.dropna(subset=['runtime'], inplace=True)
movie_summary_overview_df.dropna(subset=['release_date'], inplace=True)
movie_summary_overview_df.dropna(subset=['overview'], inplace=True)
movie_summary_overview_df['popularity'] = movie_summary_clean_df['popularity'].astype(float)
movie_summary_overview_df.info()
```

```python
movie_summary_overview_df['title'] = movie_summary_overview_df['title'].str.replace('"', '').str.replace("'", '')
```

The following is code that creates a csv file with movies that include state names or cities associated with states in both the Overview and Movie Title

```python
# iterate through movie_summary_overview_df to pull all the movies with each state in the overview
states_by_overview_df = pd.DataFrame(columns=['state', 'movie_title', 'release_date'])
for i, row in states_db.iterrows():
    current_state = row['state']
    p = '([^\w]|^)' + current_state
    movie_overviews = movie_summary_overview_df[movie_summary_overview_df['overview'].str.contains(p, case=False)].sort_values(by='popularity', ascending=False)
    for j, row_2 in movie_overviews.iterrows():
        current_title = row_2['title']
        current_release = row_2['release_date']
        new_row = pd.DataFrame({'state': [current_state], 'movie_title': [current_title], 'release_date': [current_release]})
        states_by_overview_df = pd.concat([states_by_overview_df, new_row], ignore_index=True)
print(states_by_overview_df)
```

```python
# check number of states for each
states_by_overview_df['state'].value_counts()
```

```python
# import city/state keys
city_keywords_db = pd.read_csv("resources/city_state_key.csv", quotechar='"', escapechar='\\')
city_keywords_db.head()
```

```python
# check for movies which mention cities or nicknames for each state
for i, row in city_keywords_db.iterrows():
    current_state = row['state']
    current_key = row['keyword']
    p = '([^\w]|^)' + current_key
    movie_overviews = movie_summary_overview_df[movie_summary_overview_df['overview'].str.contains(p, case=False)].sort_values(by='popularity', ascending=False)
    for j, row_2 in movie_overviews.iterrows():
        current_title = row_2['title']
        current_release = row_2['release_date']
        new_row = pd.DataFrame({'state': [current_state], 'movie_title': [current_title], 'release_date': [current_release]})
        states_by_overview_df = pd.concat([states_by_overview_df, new_row], ignore_index=True)
print(states_by_overview_df)
```

```python
# check movies by state
states_by_overview_df = states_by_overview_df.drop_duplicates()
states_by_overview_df['state'].value_counts()
```

```python
# merge the movie summary dataframe onto states_by_overview on the movie title and release date
states_by_overview_df['release_date'] = pd.to_datetime(states_by_overview_df['release_date'])
movie_summary_clean_df['release_date'] = pd.to_datetime(movie_summary_clean_df['release_date'])

merged_states_by_overview_df = pd.merge(states_by_overview_df, movie_summary_clean_df, left_on=['movie_title', 'release_date'], right_on=['title', 'release_date'])
merged_states_by_overview_df.drop(['movie_title'], axis=1, inplace=True)
merged_states_by_overview_df.head()
```

```python
# save results as csv to include in sql
merged_states_by_overview_df.rename(columns = {'state':'us_state'}, inplace = True)
merged_states_by_overview_df.to_csv('resources/only_overview/state_overview_movies_v2.csv', index=False)
merged_states_by_overview_df.info()
```

```python
# iterate through movie titles and add to table if they contain names of states
# and are not already in merged states by overview

new_movies_by_title = pd.DataFrame(columns=['us_state', 'movie_title', 'release_date'])
for i, row in states_db.iterrows():
    current_state = row['state']
    p = '([^\w]|^)' + current_state
    movie_titles = movie_summary_clean_df[movie_summary_clean_df['title'].str.contains(p, case=False)].sort_values(by='popularity', ascending=False)
    if len(movie_titles) > 0:
        for j, row_2 in movie_titles.iterrows():
            current_title = row_2['title']
            current_release = row_2['release_date']
            if merged_states_by_overview_df[(merged_states_by_overview_df['us_state'] == current_state) & (merged_states_by_overview_df['title'] == current_title) & (merged_states_by_overview_df['release_date'] == current_release)].any().any():
                print("Movie already in DataFrame")
            else:
                new_row = pd.DataFrame({'us_state': [current_state], 'movie_title': [current_title], 'release_date': [current_release]})
                new_movies_by_title = pd.concat([new_movies_by_title, new_row], ignore_index=True)
print(new_movies_by_title)
```

```python
new_movies_by_title.info()
new_movies_by_title.head(10)
```

```python
# check movie titles for state keywords
for i, row in city_keywords_db.iterrows():
    current_state = row['state']
    current_key = row['keyword']
    p = '([^\w]|^)' + current_key
    movie_titles = movie_summary_clean_df[movie_summary_clean_df['title'].str.contains(p, case=False)].sort_values(by='popularity', ascending=False)
    if len(movie_titles) > 0:
        for j, row_2 in movie_titles.iterrows():
            current_title = row_2['title']
            current_release = row_2['release_date']
            if merged_states_by_overview_df[(merged_states_by_overview_df['us_state'] == current_state) & (merged_states_by_overview_df['title'] == current_title) & (merged_states_by_overview_df['release_date'] == current_release)].any().any():
                print("Movie already in DataFrame")
            else:
                new_row = pd.DataFrame({'us_state': [current_state], 'movie_title': [current_title], 'release_date': [current_release]})
                new_movies_by_title = pd.concat([new_movies_by_title, new_row], ignore_index=True)
print(new_movies_by_title)
```

```python
# drop duplicates and check info of new movies to add
new_movies_by_title = new_movies_by_title.drop_duplicates()
new_movies_by_title.info()
new_movies_by_title.head(10)
```

```python
# merge with movie summary table and check columns
new_movies_by_title['release_date'] = pd.to_datetime(new_movies_by_title['release_date'])
movie_summary_clean_df['release_date'] = pd.to_datetime(movie_summary_clean_df['release_date'])

merged_states_by_titles_df = pd.merge(new_movies_by_title, movie_summary_clean_df, left_on=['movie_title', 'release_date'], right_on=['title', 'release_date'])
merged_states_by_titles_df.drop(['movie_title'], axis=1, inplace=True)
merged_states_by_titles_df.head()
```

```python
# Combine merged DataFrames by both overview and title
merged_titles_and_overviews_df = pd.concat([merged_states_by_overview_df, merged_states_by_titles_df], ignore_index=True)
merged_titles_and_overviews_df.info()
merged_titles_and_overviews_df.head(10)
```
